### Handling the Null values using mode, median, NAN where data dict say NAN and zero for numerical that are missing completely at random(MCAR)

```python
none_columns =  ["BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond"]
zero_columns = ["MasVnrArea","GarageYrBlt"]

for col in none_columns:
    df[col] = df[col].fillna("None")


for col in zero_columns:
    #print(col)
    df[col] = df[col].fillna(0)

median_LotFrontage = df["LotFrontage"].median()
df["LotFrontage"] = df["LotFrontage"].fillna(median_LotFrontage)

mode_Electrical = df["Electrical"].mode()[0]
df["Electrical"] = df["Electrical"].fillna(mode_Electrical)


```
